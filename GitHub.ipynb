```python
# Configure git and push from Colab
import subprocess

repo = '/content/drive/MyDrive/PK-PyAEZ/PK-PyAEZ'

cmds = [
    ['git', '-C', repo, 'config', 'user.email', 'saqibutm@outlook.com'],
    ['git', '-C', repo, 'config', 'user.name', 'Saqib'],
    ['git', '-C', repo, 'remote', 'set-url', 'origin', 'https://YOUR_USERNAME:YOUR_TOKEN@github.com/YOUR_USERNAME/PK-PyAEZ.git'],
    ['git', '-C', repo, 'push', '-u', 'origin', 'main'],
]

for cmd in cmds:
    r = subprocess.run(cmd, capture_output=True, text=True)
    print(' '.join(cmd[-2:]), '→', r.stdout.strip() or r.stderr.strip())

Replace YOUR_USERNAME and YOUR_TOKEN with your GitHub username and a Personal Access Token (PAT).
```
